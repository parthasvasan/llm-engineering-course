```python
from openai import OpenAI
import os
from dotenv import load_dotenv
from IPython.display import display, Markdown
```

```python
load_dotenv(override=True, dotenv_path="../config/.env")

ollama_base_url = os.getenv("OLLAMA_OPENAI_EP")
if not ollama_base_url:
    raise ValueError ("OLLAMA_OPENAI_EP is not set!!")
```

```python
#Solution using ollama
MODEL_NAME="phi3"
ollama_system_prompt = """
You are a helpful assistant that can read coe written in any of the programming languages and explain what the code
is doing with additional references and examples. Include reasoning behind your explanation to add clarity. If you don't know,
say so.
"""

ollama_user_prompt = """
Here is a code snippet from an application. Explain what the code block is doing and the reasoning behind your explantion

yield from {book.get("author") for book in books if book.get("author")} 
"""

messages = [
    {"role": "system", "content": ollama_system_prompt},
    {"role": "user", "content": ollama_user_prompt}
]

ollama_client = OpenAI(base_url=ollama_base_url, api_key="ollama")
response = ollama_client.chat.completions.create(
    model=MODEL_NAME,
    messages=messages
)
display(Markdown(response.choices[0].message.content))
```
