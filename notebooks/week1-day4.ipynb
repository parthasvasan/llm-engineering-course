```python
import tiktoken
```

```python
token_encoding = tiktoken.encoding_for_model("gpt-4o-mini")
tokens = token_encoding.encode("Hi, my name is bob!")
print(tokens)
```

```python
for token in tokens:
    token_text = token_encoding.decode([token])
    print (f"{token} :: {token_text}")
```

```python
for id in range(1,50):
    print(token_encoding.decode([id]))
```

```python
### Demo of how LLM's remember conversations
from openai import OpenAI
```

```python
base_url = "http://localhost:11434/v1"
api_key = "ollama"

ollama_client = OpenAI (base_url=base_url, api_key=api_key)
```

```python
MODEL_NAME="phi3"
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi there. My name is Bob."}
]
```

```python
response = ollama_client.chat.completions.create(
    model=MODEL_NAME,
    messages=messages
)

response.choices[0].message.content
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi there. What is my name?"}
]

response = ollama_client.chat.completions.create(
    model=MODEL_NAME,
    messages=messages
)

response.choices[0].message.content
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi there. My name is Bob."},
    {"role": "assistant", "content": "Hello, Bob! How may I assist you today?"},
    {"role": "user", "content": "Hi there. What is my name?"}
]

response = ollama_client.chat.completions.create(
    model=MODEL_NAME,
    messages=messages
)

response.choices[0].message.content
```
